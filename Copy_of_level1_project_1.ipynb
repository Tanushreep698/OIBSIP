```python
# ======================================
# üìä Project 1 - EDA on Retail Sales Data (Level 1)
# Google Colab Ready
# ======================================

# === 1. Install dependencies ===
!pip install pandas numpy matplotlib seaborn --quiet

# === 2. Import libraries ===
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files

plt.rcParams["figure.figsize"] = (10,6)
plt.rcParams["axes.grid"] = True

print("‚úÖ Libraries imported successfully.")

# === 3. Upload dataset ===
print("üìÇ Please upload your retail sales CSV file...")
uploaded = files.upload()

# Get uploaded file name
file_name = list(uploaded.keys())[0]
print(f"üìÑ Uploaded file: {file_name}")

# === 4. Load dataset ===
DATASET_PATH = file_name
df = pd.read_csv(DATASET_PATH)
print(f"‚úÖ Data loaded successfully. Shape: {df.shape}")

# === 5. Preview data ===
display(df.head())
display(df.tail())

# === 6. Basic info ===
print("\n--- INFO ---")
print(df.info())

print("\n--- DESCRIBE (numeric) ---")
display(df.describe().T)

print("\n--- Missing Values ---")
missing = df.isna().sum()
print(missing[missing > 0])

# === 7. Basic Cleaning ===
# Strip strings
for col in df.select_dtypes(include="object").columns:
    df[col] = df[col].astype(str).str.strip()

# Convert date columns
for col in df.columns:
    if "date" in col.lower():
        df[col] = pd.to_datetime(df[col], errors="coerce")

# Convert numeric columns
for col in df.columns:
    if col.lower() in ["sales", "amount", "price", "unitprice", "quantity", "qty"]:
        df[col] = pd.to_numeric(df[col], errors="coerce")

print("‚úÖ Basic cleaning done.")

# === 8. Time Series Analysis ===
date_cols = [c for c in df.columns if 'date' in c.lower()]
if date_cols:
    dt_col = date_cols[0]
    value_col = None
    for c in ["sales","amount","revenue"]:
        if c in map(str.lower, df.columns):
            value_col = c
            break

    if not value_col:
        value_col = df.select_dtypes(include=np.number).columns[0]

    ts = df.groupby(df[dt_col].dt.to_period("M"))[value_col].sum()
    ts.index = ts.index.to_timestamp()

    ts.plot(marker="o")
    plt.title(f"üìà Monthly {value_col} Over Time")
    plt.xlabel("Month")
    plt.ylabel(value_col.capitalize())
    plt.show()
else:
    print("‚ö†Ô∏è No date column found for time series.")

# === 9. Top Customers ===
for cust_col in ["Customer", "CustomerID", "Customer ID", "customer_id"]:
    if cust_col in df.columns:
        df[cust_col].value_counts().head(10).plot(kind="bar", color="skyblue")
        plt.title("üèÜ Top 10 Customers by Orders")
        plt.show()
        break

# === 10. Top Products ===
for prod_col in ["Product", "ProductName", "Product Name", "Item", "SKU"]:
    if prod_col in df.columns:
        df[prod_col].value_counts().head(10).plot(kind="bar", color="orange")
        plt.title("üì¶ Top 10 Products by Orders")
        plt.show()
        break

# === 11. Correlation Heatmap ===
num_df = df.select_dtypes(include=np.number)
if not num_df.empty:
    sns.heatmap(num_df.corr(), annot=True, cmap="coolwarm")
    plt.title("üîç Correlation Heatmap")
    plt.show()

# === 12. Insights & Recommendations ===
print("\nüí° Insights & Recommendations:")
print("- Identify seasonal sales peaks and prepare inventory/marketing campaigns.")
print("- Focus on top products and customers for loyalty programs.")
print("- Investigate low-performing categories for improvement.")
print("- Monitor trends for pricing optimization.")
```
