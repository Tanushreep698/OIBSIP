```python
# ðŸ“¦ Step 1: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files

# ðŸ“‚ Step 2: Upload and Load Dataset
uploaded = files.upload()
df = pd.read_csv(next(iter(uploaded)))

# ðŸ§­ Step 3: Basic Overview
print("Shape:", df.shape)
print("Columns:", df.columns.tolist())
print("\nQuality Value Counts:\n", df['quality'].value_counts())

# ðŸ“Š Step 4: Distribution of Wine Quality
plt.figure(figsize=(8,5))
sns.countplot(x='quality', data=df, palette='viridis')
plt.title("Wine Quality Distribution")
plt.xlabel("Quality Score")
plt.ylabel("Count")
plt.show()

# ðŸ“ˆ Step 5: Correlation Heatmap
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()

# ðŸ“‰ Step 6: Alcohol vs Quality
plt.figure(figsize=(8,5))
sns.boxplot(x='quality', y='alcohol', data=df, palette='magma')
plt.title("Alcohol Content by Wine Quality")
plt.show()

# ðŸ“Š Step 7: Sulphates vs Quality
plt.figure(figsize=(8,5))
sns.boxplot(x='quality', y='sulphates', data=df, palette='cool')
plt.title("Sulphates by Wine Quality")
plt.show()

# ðŸ“Š Step 8: Volatile Acidity vs Quality
plt.figure(figsize=(8,5))
sns.boxplot(x='quality', y='volatile acidity', data=df, palette='cubehelix')
plt.title("Volatile Acidity by Wine Quality")
plt.show()

# ðŸ§  Step 9: Insights
print("\nTop correlations with quality:")
print(df.corr()['quality'].sort_values(ascending=False)[1:6])
```
